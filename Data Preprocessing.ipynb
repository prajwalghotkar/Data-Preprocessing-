# Data Manipulation:

## Q 1) Create a new column -’total_benefitted’ that is a sum of inmates benefitted through all modes.


```python
prisoners["Total_Benefited"]=prisoners.iloc[:,2:].sum(axis=1)
display(prisoners)
```

## Q 2) Create a new Data Frame to Store column-wise total. Name New Data Frame as prisoners_total 

```python
prisoners_total=prisoners
display(prisoners_total)
```

## Q 3) Create a new row - “Prisoners Total'” that is the sum of all inmates benefitted through each mode across all states.

###  Change the data type of the year column to String from integer. Since there is no addition for the YEAR column, NaN will be populated to this column and the data type changes from integer to float automatically. To avoid this explicitly change the data type to string. 


```python
prisoners_total['YEAR']=prisoners_total['YEAR'].astype('str')
# Add column-wise total where the data type is numeric
prisoners_total.loc['Prisoners Total']=prisoners_total.iloc[:,2:].sum(numeric_only=True)
display(prisoners_total)
```

## Q 4) First two columns will be populated as NaN, Replace the NaN with empty string for better formatting.

```python
prisoners_total.loc[prisoners_total.index[-1],['STATE/UT','YEAR']]=''
display(prisoners_total)
```

# Plotting:

## Q 1) Make a bar plot with each state name on the x -axis and their total benefitted inmates as their bar heights. Which state has the maximum number of beneficiaries?

```python
import numpy as np
xlabels = prisoners['STATE/UT'].values
plot.figure(figsize=(20, 10))

plot.xticks(np.arange(xlabels.shape[0]),xlabels,rotation='vertical', fontsize=18)
plot.xticks

plot.bar(np.arange(prisoners.iloc[:-1].values.shape[0]),prisoners['Total_Benefited'].iloc[:-1],align='edge')
plot.show()
```

## Q 2) Make a pie chart that depicts the ratio among different modes of benefits.

```python
labels = prisoners.columns[2:-1]
sizes = prisoners_total.iloc[-1][2:-1].values
fig1, ax1 = plot.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plot.show()
```
